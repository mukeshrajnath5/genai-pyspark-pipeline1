```python
import random
from faker import Faker
from IPython.display import HTML, display

fake = Faker()

# 1. The 10 Compatibility Questions
QUESTIONS = [
    "I'd rather stay in and watch a movie than go to a crowded party.",
    "I prefer texting over talking on the phone.",
    "I like having a plan rather than just figuring things out as we go.",
    "I'm always down to try weird or unusual foods.",
    "When I'm upset, I usually need some space before I can talk about it.",
    "I'm definitely more of a saver than a spender.",
    "I spend way too much time sending memes and reels to my friends.",
    "My room or apartment is usually a little messy, and I'm okay with that.",
    "Having pets in the house is an absolute must for me.",
    "I secretly love cheesy romantic gestures."
]

# 2. Hinge-Style Prompts
PROFILE_PROMPTS = [
    "My ideal weekend...",
    "I geek out on...",
    "A boundary of mine is...",
    "Must love...",
    "The quickest way to my heart is...",
    "Typical Sunday..."
]

# 3. Generate Mock Database
def generate_mock_database(num_users=100):
    database = []
    genders = ['Male', 'Female']

    for user_id in range(1, num_users + 1):
        # Select 2 random prompts for their profile
        selected_prompts = random.sample(PROFILE_PROMPTS, 2)

        profile = {
            "id": user_id,
            "name": fake.first_name(),
            "age": random.randint(18, 26),
            "gender": random.choice(genders),
            "location": fake.city(),
            # pravatar gives us realistic human faces based on an ID number
            "photo_url": f"https://i.pravatar.cc/600?img={random.randint(1, 70)}",
            "prompts": [
                {"question": selected_prompts[0], "answer": fake.sentence(nb_words=6)},
                {"question": selected_prompts[1], "answer": fake.sentence(nb_words=8)}
            ],
            "answers": [random.randint(1, 5) for _ in range(10)]
        }
        database.append(profile)
    return database

# 4. The Matching Algorithm
def calculate_compatibility(user_answers, target_answers):
    total_difference = sum(abs(u - t) for u, t in zip(user_answers, target_answers))
    max_diff = len(user_answers) * 4
    compatibility_percentage = ((max_diff - total_difference) / max_diff) * 100
    return round(compatibility_percentage, 2)

def find_matches(new_user_answers, database, top_n=3):
    matches = []
    for profile in database:
        score = calculate_compatibility(new_user_answers, profile["answers"])
        matches.append({"profile": profile, "compatibility_score": score})

    matches.sort(key=lambda x: x["compatibility_score"], reverse=True)
    return matches[:top_n]

# 5. The Hinge/Bumble Inspired UI Visualizer
def display_matches_html(matches):
    # This CSS creates a simulated "phone screen" layout for the matches
    html_content = """
    <div style='display: flex; gap: 30px; padding: 20px; overflow-x: auto; background: #f4f5f6; font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Roboto, Helvetica, Arial, sans-serif;'>
    """

    for match in matches:
        p = match["profile"]
        score = match["compatibility_score"]

        # Build the individual profile card
        card_html = f"""
        <div style="min-width: 340px; max-width: 340px; background: white; border-radius: 16px; box-shadow: 0 8px 24px rgba(0,0,0,0.08); overflow: hidden; position: relative;">
            <div style="position: relative; height: 400px;">
                <img src="{p['photo_url']}" style="width: 100%; height: 100%; object-fit: cover;">
                <div style="position: absolute; bottom: 0; left: 0; right: 0; padding: 60px 20px 20px; background: linear-gradient(transparent, rgba(0,0,0,0.8)); color: white;">
                    <h2 style="margin: 0; font-size: 28px; font-weight: 700;">{p['name']}, {p['age']}</h2>
                    <p style="margin: 4px 0 0; font-size: 15px; opacity: 0.9;">📍 {p['location']}</p>
                </div>
                <div style="position: absolute; top: 16px; right: 16px; background: rgba(0,0,0,0.6); backdrop-filter: blur(4px); color: white; padding: 6px 12px; border-radius: 20px; font-weight: 600; font-size: 14px; border: 1px solid rgba(255,255,255,0.2);">
                    {score}% Match
                </div>
            </div>

            <div style="padding: 24px 20px; background: white;">
                <div style="margin-bottom: 20px;">
                    <p style="font-size: 13px; color: #707070; margin: 0 0 6px 0; font-weight: 600; text-transform: uppercase; letter-spacing: 0.5px;">{p['prompts'][0]['question']}</p>
                    <p style="margin: 0; font-size: 20px; color: #111; font-family: Georgia, serif; line-height: 1.4;">"{p['prompts'][0]['answer']}"</p>
                </div>

                <div style="height: 1px; background: #eee; margin: 20px 0;"></div>

                <div>
                    <p style="font-size: 13px; color: #707070; margin: 0 0 6px 0; font-weight: 600; text-transform: uppercase; letter-spacing: 0.5px;">{p['prompts'][1]['question']}</p>
                    <p style="margin: 0; font-size: 20px; color: #111; font-family: Georgia, serif; line-height: 1.4;">"{p['prompts'][1]['answer']}"</p>
                </div>
            </div>
        </div>
        """
        html_content += card_html

    html_content += "</div>"
    display(HTML(html_content))

# 6. Run the Application
def run_dating_app():
    print("=== Mindset Matchmaker Backend Prototype ===\n")
    db = generate_mock_database(150)

    # Injecting a specific test profile right before the matching process
    # to demonstrate how the backend handles highly specific lifestyle matches.
    test_profile = {
        "id": 999,
        "name": "Aditi",
        "age": 22,
        "gender": "Female",
        "location": "Chennai",
        "photo_url": "https://i.pravatar.cc/600?img=44",
        "prompts": [
            {"question": "My ideal weekend...", "answer": "Taking the Royal Enfield Bullet 350 out for a ride around the city."},
            {"question": "Must love...", "answer": "Dogs! Especially sweet Pitbull mixes. And don't judge how much I play GTA."}
        ],
        "answers": [] # Will be filled dynamically
    }
    db.append(test_profile)

    print("Answer the following on a scale of 1 (Strongly Disagree) to 5 (Strongly Agree):\n")

    user_answers = []
    for i, q in enumerate(QUESTIONS):
        while True:
            try:
                ans = int(input(f"Q{i+1}: {q}\nYour Answer (1-5): "))
                if 1 <= ans <= 5:
                    user_answers.append(ans)
                    break
                print("Invalid input. Please enter a number between 1 and 5.")
            except ValueError:
                print("Invalid input. Please enter a valid integer.")
        print("-" * 40)

    # Set the test profile answers to perfectly match the user input to test the 100% logic
    test_profile["answers"] = user_answers.copy()

    print("\nCalculating matches...\n")
    top_matches = find_matches(user_answers, db, top_n=3)

    display_matches_html(top_matches)

if __name__ == "__main__":
    run_dating_app()
```
